Top 10 States with highest sales

```python
sales_by_state = df1.groupby('State')['Sales'].sum().sort_values(ascending=False)

# Select top 10 states with highest sales
top_10_states = sales_by_state.head(10)

# Create a bar plot
plt.figure(figsize=(10, 6))
top_10_states.plot(kind='bar', color='skyblue')
plt.title('Top 10 States with Highest Sales')
plt.xlabel('State')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
df1['Order Date'] = pd.to_datetime(df1['Order Date'])

# Group by 'Order Date' and calculate total sales for each date
sales_over_time = df1.groupby('Order Date')['Sales'].sum()

# Create a time series plot
plt.figure(figsize=(12, 6))
sales_over_time.plot(color='green')
plt.title('Sales Over Time')
plt.xlabel('Order Date')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()
```

```python
df1['Order Date'] = pd.to_datetime(df1['Order Date'])

# Group by 'Order Date' and 'Category' and calculate total sales for each combination
sales_by_category_date = df1.groupby(['Order Date', 'Category'])['Sales'].sum().unstack()

# Plot time series for each product category
plt.figure(figsize=(12, 6))
for category in sales_by_category_date.columns:
    plt.plot(sales_by_category_date.index, sales_by_category_date[category], label=category)

plt.title('Time Series Analysis of Sales by Product Category')
plt.xlabel('Order Date')
plt.ylabel('Total Sales')
plt.legend(title='Product Category', loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True)
plt.show()
```

```python
class EDA:
    def __init__(self, df1, df2):
        self.df1 = df1
        self.df2 = df2

    def explore_data(self, *args, **kwargs):
        try:
            for arg in args:
                if arg == 'summary':
                    self.summary()
                elif arg == 'shape':
                    self.shape()
                # Add more functionality here based on arguments

        except Exception as e:
            print(f"An error occurred during EDA: {e}")

    def summary(self):
        print("Summary of DataFrame 1:")
        print(self.df1.describe())
        print("\nSummary of DataFrame 2:")
        print(self.df2.describe())

    def shape(self):
        print("Shape of DataFrame 1:", self.df1.shape)
        print("Shape of DataFrame 2:", self.df2.shape)

```

```python
def custom_function(*args, **kwargs):
    try:
        # Example of using a class method
        eda = EDA(kwargs.get('df1'), kwargs.get('df2'))
        eda.explore_data(*args)

    except Exception as e:
        print(f"An error occurred in the custom function: {e}")

# Example usage of the custom function
df1 = pd.DataFrame({'Sales': [100, 200, 300], 'Quantity': [1, 2, 3]})
df2 = pd.DataFrame({'Profit': [50, 100, 150], 'Discount': [0.1, 0.2, 0.3]})

custom_function('summary', 'shape', df1=df1, df2=df2)
```
